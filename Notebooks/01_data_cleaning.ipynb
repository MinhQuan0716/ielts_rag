```python
import pandas as pd
```

```python
file_path= '../Data/raw/ielts_writing_dataset.csv'
```

```python
# 1. Load the dataset
# Replace 'ielts_essays.csv' with the actual path to your file
try:
    df = pd.read_csv(file_path)
    print("Dataset loaded successfully!\n")
except FileNotFoundError:
    print(f"Error: Could not find {file_path}. Please check the file name and path.")
```

```python
#2 Initial Exploration
print("--- Basic Dataset Information ---")
print(f"Total rows (essays): {df.shape[0]}")
print(f"Total columns: {df.shape[1]}")
print("\nColumn Names:")
print(df.columns.tolist())
print("\n--- First 3 Rows ---")
display(df.head(3))
```

```python
# 3. Data Cleaning
print("\n--- Starting Data Cleaning ---")

# Standardize column names (makes them lowercase and replaces spaces with underscores)
df.columns = df.columns.str.lower().str.replace(' ', '_')

# A. Drop rows missing the actual essay text or the overall score
# (Adjust 'essay_text' and 'overall_score' to match your actual column names)
text_col = 'essay' # Change this if your column is named 'essay' or 'text'
score_col = 'overall' # Change this if your column is named 'score' or 'band'

if text_col in df.columns and score_col in df.columns:
    initial_count = len(df)
    df = df.dropna(subset=[text_col, score_col])
    dropped_count = initial_count - len(df)
    print(f"Dropped {dropped_count} rows due to missing essays or scores.")
else:
    print(f"Warning: Could not find columns '{text_col}' or '{score_col}'. Please check your column names.")

# B. Filter for Task 2 only (if your dataset includes Task 1)
# (Adjust 'task_type' to match your column name, e.g., 'type', 'question_type')
task_col = 'task_type'
if task_col in df.columns:
    # Assuming Task 2 is represented by the number 2 or string 'Task 2'
    df = df[df[task_col].astype(str).str.contains('2')]
    print(f"Filtered for Task 2. Remaining essays: {len(df)}")
```

```python
# 4. Checking for Sub-scores (The Goldmine for your AI)
print("\n--- Checking for Sub-scores ---")
# List of common names for the 4 grading criteria
sub_score_keywords = ['task_response', 'coherence_cohesion', 'lexical_resource', 'grammar', 'tr', 'cc', 'lr', 'gra']

found_sub_scores = [col for col in df.columns if any(keyword in col for keyword in sub_score_keywords)]

if found_sub_scores:
    print(f"Great! Found potential sub-score columns: {found_sub_scores}")
    # Let's see the distribution of overall scores
    print("\nOverall Score Distribution:")
    print(df[score_col].value_counts().sort_index())
else:
    print("Could not clearly identify sub-score columns (TR, CC, LR, GRA).")
    print("If your dataset only has overall scores, your AI's feedback will be less detailed.")
```

```python
# 5. Save the cleaned dataset
cleaned_file_path = '../Data/processed/cleaned_ielts_task2_essays.csv'
df.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned dataset saved to: {cleaned_file_path}")
```
